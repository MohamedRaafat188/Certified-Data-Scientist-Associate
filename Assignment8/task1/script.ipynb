```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize': [15, 15]}, font_scale=1.3)
```

```python
df = pd.read_csv('sendy_logistics.csv')
df
```

```python
def extact_userid(id):
    index = id.rfind('_')
    return id[index + 1:]
```

```python
df['USER_ID'] = df['User Id'].apply(extact_userid)
df
```

```python
df = pd.read_csv('titanic_full.csv')
df
```

```python
def extract_title(name):
    start_index = name.find(", ") + 2
    end_index = name.find(". ")
    return name[start_index: end_index]
```

```python
df['Title'] = df['Name'].apply(extract_title)
df
```

```python
sns.countplot(x='Title', data=df)
```

* number of people have the title 'Mr' is almost 515
* number of people have the title 'Mrs' is almost 130
* number of people have the title 'Miss' is almost 185
* number of people have the title 'Master' is almost 40
