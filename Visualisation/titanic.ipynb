```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.drop(columns=['pclass', 'sibsp', 'parch', 'embarked', 'adult_male', 'deck'], inplace=True)
df
```

```python
sns.displot(df['age'])
```

* most people on the ship were in youth age from 16 to 36 years old

```python
sns.kdeplot(df['fare'])
```

* almost all people paid fare less than 100 pound

```python
sns.displot(df['fare'], bins=25)
```

```python
sns.displot(df['survived'])
```

* number of dead people is about 1.5 of number of survived people

```python
df['survived'].value_counts()
```

```python
sns.jointplot(x='survived', y='age', data=df)
```

* there is no correlation between age and surviving

```python
sns.jointplot(x='survived', y='fare', data=df, kind='kde')
```

* there is no correlation between fare and surviving

* there is no correlation between fare and age

```python
df
```

```python
sns.countplot(x='sex', data=df)
```

* number of males is almost twice the number of females

```python
sns.countplot(x='class', data=df, palette='Set1')
```

* most people were third class and first class was slightly bigger than second class

```python
sns.countplot(x='who', data=df, palette='Set1_r')
```

* there ware 60% men, 30% women, and 10% children

```python
sns.countplot(x='embark_town', data=df, palette='turbo')
```

* most people were from Southampton and few people from the other two cities

```python
sns.countplot(y='alive', data=df, palette='dark')
```

* Almost a third of the people survived

```python
sns.countplot(y='alone', data=df, palette='Paired_r')
```

* people who travelled alone is about 1.5 of people who travelled with their families

```python
sns.swarmplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df, hue='alive')
```

* most men in the age from 17 to 43
* most women in the age from 14 to 40
* most dead people were from men 

```python
sns.boxplot(x='sex', y='fare', data=df, hue='embark_town')
```

people from cherbourg paid large fares

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='class', y='age', data=df)
plt.subplot(1, 2, 2)
sns.boxplot(x='class', y='fare', data=df)
```

* the age increases in the first class
* there are strange data such as there are people in the first class paid fare like people in the third class

```python
sns.violinplot(x='who', y='fare', data=df, hue='alive', split=True)
```

* fare doesn't depend on who the person is
* the percentage of dead people is highest in men and lowest in children and that is reversed in the survived people

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='embark_town', y='age', data=df, hue='alone')
plt.subplot(1, 2, 2)
sns.boxplot(x='embark_town', y='fare', data=df, hue='alone')
```

* there is no relation between the wown and the age
* people from cherbourg paid more

```python
plt.subplot(1, 2, 1)
sns.swarmplot(x='alive', y='age', data=df)
plt.subplot(1, 2, 2)
sns.boxplot(x='alive', y='fare', data=df)
```

* no correlation between age and surviving
* the more the people paid the more they have chance to survive

```python
plt.subplot(1, 2, 1)
sns.swarmplot(x='alone', y='age', data=df)
plt.subplot(1, 2, 2)
sns.swarmplot(x='alone', y='fare', data=df)
```

* the people who travelled alone were in the age from 16 to 40
* people who travelled alone paid less than other people

the people from southampton paid the most despite low average in their fare and that proofs that the most people in the ship were from southampton
